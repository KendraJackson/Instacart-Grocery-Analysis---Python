```python
# Import Libraries
import pandas as pd
import numpy as np
import os
```

### Step 3. Import the Data

```python
# Import Data
# turn folders into string
path = r'C:\Users\Jamie&Ken\Desktop\Data Analytics Course\Python Fundeamentals for Data Analysts\08-2024 Instacart Basket Analysis'
# Import products file
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
# Import orders combined file
df_ords_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

### Step 5. Merge df_prods with df_ords_combined 

```python
#Drop unnamed columns in df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# Drop unnamed columns in df_ords_combined (due to one being exported as excel file)
df_ords_combined = df_ords_combined.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

```python
df_merged = df_prods.merge(df_ords_combined, on = ['product_id'], indicator = True)
```

```python
df_merged.head()
```

### Step 6. Confirm the results of the merge using merge flag

```python
# Checking for full merge without overwriting
df_merged['_merge'].value_counts()
```

```python
# Checking Shape
df_merged.shape
```

### Step 7. Export the file

```python
# Export orders_products_merged filed
df_merged.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```
