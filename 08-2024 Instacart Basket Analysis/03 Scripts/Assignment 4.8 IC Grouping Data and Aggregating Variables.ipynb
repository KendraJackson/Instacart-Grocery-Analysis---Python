#### 2. Finding mean of order_number column by department_id grouping

```python
# Step 2. Group data by department_id and aggregate to obtain mean values
ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

#### 4. Create a Loyalty Flag

```python
# Create group and aggregate with transform (new column "max_order')
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
# Checking results
pd.options.display.max_rows = None
ords_prods_merge.head(10)
```

```python
# Create new column (flag) for loyalty
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check code execution
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number', 'max_order']].head(10)
```

#### 5/6. Defining Customer Spending

```python
# Step 5 the min, max, and mean of each loyalty flag
ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean', 'min', 'max', 'sum']})
```

###### Results: Loyal customer seem to spend on product prices with lesser value while new customers tend to buy more expensive products. Overall, regular customers spend the most money with loyal and then new customers coming after.  

```python
# Step 6 Create column for average spender
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Create flag for low and high spender
ords_prods_merge.loc[ords_prods_merge['average_spend'] <10, 'spender_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['average_spend'] >=10, 'spender_flag'] = 'High spender'
```

```python
# Check flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Check code executed correctly
ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(50)
```

#### 7. Defining frequent customers

```python
# Create column for "customer frequency"
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Create flag on order frequency
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['customer_frequency'] > 10) & (ords_prods_merge['customer_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency 
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Check code execution
ords_prods_merge[['user_id', 'days_since_prior_order', 'customer_frequency', 'frequency_flag']].head(100)
```
