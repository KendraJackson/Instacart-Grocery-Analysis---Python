### 2. Creating price_label column

```python
# Defining new function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# Using Loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

### 4. Update busiest day column (Assignment Start)

```python
# find busiest and slowest days
ords_prods_merge['weekday_of_order'].value_counts(dropna = False)
```

###### busiest days 0 - Saturday, 1 Sunday Slowest days 4 -Wednesday, 3 - Tuesday

```python
# create if statement with for-loop
result_2 = []

for value in ords_prods_merge["weekday_of_order"]:
    if value == 0 or value == 1:
        result_2.append("Busiest days")
    elif value == 4 or value == 3:
        result_2.append("Slowest days")
    else:
        result_2.append("Average days")
```

### 6. Create a new column busiest day period

```python
# Find busiest and slowest hours
ords_prods_merge['order_hour_of_day'].value_counts()
```

##### Splitting the above into categories of equal 3rds. Most orders = 10,11,14,15,13,12,16,9 Fewest orders = 23,6,0,1,5,2,4,3 Average orders = all remaining values

```python
# Create if statement using a for-loop labelling periods of time as "Most orders", "Average orders", and "Fewest Orders"

result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_3.append("Most orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_3.append("Fewest orders")
    else:
        result_3.append("Average orders")
```

```python
# Check result
result_3
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge
ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
# Print Frequency
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
